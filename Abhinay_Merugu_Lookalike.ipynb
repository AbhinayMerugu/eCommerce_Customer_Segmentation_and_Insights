```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

```python
final_df=pd.read_csv("/content/final_df")
```

```python
final_df.head()
```

```python
cosine_sim = cosine_similarity(final_df)
```

```python
recommendation_dict = {}
for id in range(20):
    user_index = id
    similarity_scores = list(enumerate(cosine_sim[user_index]))
    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)
    top_3_recommendations = similarity_scores[1:4]
    recommendation_dict[id+1] = [(final_df.loc[x[0],"CustomerID"], x[1]) for x in top_3_recommendations]
```

```python
recommendation_data = []
for user_id, recommended_users in recommendation_dict.items():
    for recommended_user in recommended_users:
        recommendation_data.append([user_id, recommended_user[0], recommended_user[1]])
```

```python
recommendation_df = pd.DataFrame(recommendation_data, columns=['CustomerID', 'Recommended_CustomerID', 'Similarity_Score'])
recommendation_df.to_csv('Abhinay_Merugu_Lookalike.csv', index=False)
```
